```python
import numpy as np
import pandas as pd
```

```python
DIR = "20210227_195345"
MODEL = "VARIANCE_GAMMA"
```

```python
path = f'_output/calibration/results/{DIR}/{MODEL}/parameters.csv'
```

```python
df = pd.read_csv(path)
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df = df.set_index('date')
#df = df.loc[:, df.columns[3:]]
df['C'] = df['NU'].apply(lambda x: 1/x)
```

```python
df
```

```python
df1 = df[(df.index >= '2019-04-01') & (df.index < '2019-10-01')]
df2 = df[(df.index >= '2019-10-01') & (df.index < '2020-02-01')]
df3 = df[(df.index >= '2020-02-01') & (df.index < '2020-07-01')]
```

```python
df1_sum = df1.mean()
df2_sum = df2.mean()
df3_sum = df3.mean()
```

```python
df1_sum
```

```python
#for BS
df_frames = [df1_sum,df2_sum,df3_sum]
df = pd.DataFrame(df_frames, index=['bullish', 'calm', 'COVID'])
columns = df.columns
df = pd.DataFrame(np.round(df_frames, 2), index=['bullish', 'calm', 'covid'], columns=columns)
#del df['count']
df
```
